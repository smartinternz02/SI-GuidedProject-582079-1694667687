```python
from sklearn.cluster import KMeans
```

```python
new_df=df[['Annual Income (k$)', 'Spending Score (1-100)']]
new_df.head()
```

```python
error= []
for i in range(1, 10):
    km = KMeans(n_clusters =  i)
    km.fit(new_df)
    error.append(km.inertia_)
```

```python
error
```

```python
plt.xlabel('Number of clusters')
plt.ylabel("Error")
plt.plot(range(1,10), error)
```

```python
km = KMeans(n_clusters = 5)
predicted = km.fit_predict(new_df)
predicted
```

```python
df['Cluster'] = predicted
df
```

```python
df1 = df[df.Cluster==0] 
df2 = df[df.Cluster==1] 
df3 = df[df.Cluster==2]
df4 = df[df.Cluster==3]
df5 = df[df.Cluster==4]
plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],color='green')
plt.scatter(df2['Annual Income (k$)'],df2['Spending Score (1-100)'],color='red')
plt.scatter(df3['Annual Income (k$)'],df3['Spending Score (1-100)'],color='black')
plt.scatter(df4['Annual Income (k$)'],df4['Spending Score (1-100)'],color='c')
plt.scatter(df5['Annual Income (k$)'],df5['Spending Score (1-100)'],color='blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
```

```python
plt.figure(figsize=(12,8))
pal = ["red","blue", "green","yellow"]
pl = sns.countplot(x=df["Cluster"], palette= pal)
pl.set_title("Distribution Of The Clusters")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=df, x='Gender', y='Spending Score (1-100)', hue='Cluster', palette= pal)
plt.title('Plot relation between Gender, Spending score (1-100) based on cluster')
plt.xlabel('Gender')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=df, x='Gender', y='Annual Income (k$)', hue='Cluster', palette= pal)
plt.title('Plot relation between Gender, Annual Income (k$) based on cluster')
plt.xlabel('Gender')
plt.ylabel('Annual Income (k$)')
plt.show()
```

# From above we can see 
Cluster 0:- It doesnt vary in Annual Income and Spending Score (1-100) with respect to Gender column

Cluster 1:- It has highest Annual Income and Spending Score (1-100) with respect to Gender Column as compared to               other clusters 

Cluster 2:- It has seen the difference between Male and Female with the values of Annual Income[low values] and               Spending Score (1-100)[high values] while females has high values than males.

Cluster 3:- It has high Spending Score (1-100) and low Annual Income with almost equal amount between male and female.

Cluster 4:- It doesnt vary in Annual Income and Spending Score (1-100) with respect to Gender column
